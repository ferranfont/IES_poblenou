```python
import numpy as np

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)
df
```

```python

df[(df['animal'] == 'cat') & (df['age'] < 3)]    # SELECT TWO CONDITIONS USING  BRACKETS
```

```python
df.loc['k'] = [5.5, 'dog', 'no', 2]     # APPEND A ROW IN K
df
```

```python
df.drop('k')                          # DROP COLUMN K
```

```python
df.sort_values(by=['age', 'visits'], ascending=[False, True])  # SORT DATAFRAME BY AGE AND VISITS, FIRST DESC, SECOND ASCE
```

```python
df['priority'] = df['priority'].map({'yes': True, 'no': False}) # FOR ALL VALUES IN COLUMN priority with yes, replace with True
```

```python
df['animal'] = df['animal'].replace('snake', 'python')   # REPLACE IN COLUM animal ALL snake with python
df
```
